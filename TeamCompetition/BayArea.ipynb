```python
import json 
import pandas as pd 
from pandas.io.json import json_normalize #package for flattening json in pandas df

#load json object
df = pd.read_json('bayarea.json', lines=True)
df
```

```python
experience_data = json_normalize(data=df['experience'][500])
print(experience_data['title'][0])
print(experience_data['org'][0])
```

```python
name_data = json_normalize(data=df['name'][50040])
name_data
```

```python
latest_title=[]
latest_org=[]
rows = df.shape[0]
for i in range(0, rows):
    if isinstance(df['experience'][i], list):
        experience_data = json_normalize(data=df['experience'][i])
        latest_title.append(experience_data['title'][0])
        latest_org.append(experience_data['org'][0])
    else:
        latest_title.append(None)
        latest_org.append(None)
df['latest_title'] = latest_title
df['latest_org'] = latest_org
```

```python
family_name=[]
given_name=[]

for i in range(0, rows):
    if isinstance(df['name'][i], dict):
        name_data = json_normalize(data=df['name'][i])
        family_name.append(name_data['family_name'][0])
        given_name.append(name_data['given_name'][0])
    else:
        family_name.append(None)
        given_name.append(None)
df['family_name'] = family_name
df['given_name'] = given_name
```

```python
df = df.drop('also_view', axis=1)
df = df.drop('events', axis=1)
df = df.drop('experience', axis=1)
df = df.drop('name', axis=1)
df = df.drop('homepage', axis=1)
df = df.drop('overview_html', axis=1)
df = df.drop('skills', axis=1)
df = df.drop('specilities', axis=1)
df = df.drop('summary', axis=1)
df = df.drop('url', axis=1)
df = df.drop('interests', axis=1)
df = df.drop('interval', axis=1)
df = df.drop('honors', axis=1)
```

```python
df
```

```python
df = df.dropna(subset=['latest_title', 'latest_org','family_name', 'given_name'])
```
